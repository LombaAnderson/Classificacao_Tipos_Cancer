# Importando as bibliotecas padrão

```python
# Bibliotecas padrão
import pandas as pd 
import numpy as np 

# Visualização de dados
import matplotlib.pyplot as plt
import seaborn as sns

# Machine Learning
# Análisa a acurácia dos modelos
from sklearn import metrics 

# Ocultando Warnings indesejados
import warnings
warnings.filterwarnings('ignore')
```

# Importando base de dados

```python
# Base de dados nativa do sklearn
from sklearn.datasets import load_breast_cancer 
```

```python
# Carregando a base de dados
dados = load_breast_cancer()
```

```python
# Descrição da base de dados
print(dados.DESCR)
```

```python
# Transformando a base de dados em um DataFrame
cancer=pd.DataFrame(data=dados.data, columns= dados.feature_names)

# Adicionando a coluna classe(matriz dependente)
cancer['Class']=dados.target
```

```python
# Salvando base de dados em csv
cancer.to_csv = ('cancer.csv')
```

```python
# Distribuição das classes
cancer['Class'].value_counts() # 1- Benigno 0 = Maligno
```

```python
# Criando gráfico -Cores do gráfico pizza
colors=['#35b2de', '#ffcb5a']

labels = cancer['Class'].value_counts().index
plt.pie(cancer['Class'].value_counts(),autopct='%1.1f%%',colors=colors) # conta as ocorrências de cada classe e exibe a porcentagem
plt.legend(labels,bbox_to_anchor=(1.25,1),) #  Legendas
plt.title('Porcentagem: Benignos x Malignos ')
plt.show()
```

```python
# Criação de amostras para a construção dos modelos
from sklearn.model_selection import train_test_split

# primeiro vamos dividir a base de dados entre matriz de recursos e matriz dependente(classe)
X= cancer.iloc[:,0:-1]# Selecionando todas as linhas, da primeira coluna até a penúltima coluna.
Y=cancer.iloc[:,-1] # Selecionando todas as linhas da última coluna ['Class'].


x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.30,random_state=42)
# test-size: neste caso vamos dividir nosso dataset em 70% treino e 30% teste
# random_state: vamos selecionar de forma aleatória
```

```python
#  bases de dados para treino e testes 
print('X treino',x_train.shape)
print('X test',x_test.shape)
print('Y treino',y_train.shape)
print('Y test',y_test.shape)
```

```python
# importando modelo de 
from sklearn.tree import DecisionTreeClassifier

tree=DecisionTreeClassifier() # Criando o modelo
tree.fit(x_train,y_train) # Treinando o modelo
y_pred=tree.predict(x_test) # predizendo
acc_tree=round(metrics.accuracy_score(y_pred,y_test)*100,1) # avaliando a acurácia. previsões x resultados reais
print(acc_tree,"% de acurácia")
```
